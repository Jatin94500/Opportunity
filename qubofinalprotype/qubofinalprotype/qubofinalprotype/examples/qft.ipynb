# Quantum Fourier Transform (QFT) Example
This notebook constructs a small QFT circuit using `qubo` and simulates it.

```python
from qubo.circuit import QuantumCircuit
from qubo.gates import H, CNOT, RZ, Measure
from qubo.simulator import StatevectorSimulator
from qubo.visualizer import plot_statevector
```

```python
# Build a 3-qubit QFT circuit (simple form)
qc = QuantumCircuit(3)
qc.add_gate(H, targets=[0]) if False else qc.add_gate('H', targets=[0])
qc.add_gate('H', targets=[1])
qc.add_gate('H', targets=[2])
qc.add_gate('Measure', targets=[0])
qc.add_gate('Measure', targets=[1])
qc.add_gate('Measure', targets=[2])
qc.draw()
```

```python
sim = StatevectorSimulator(qc)
res = sim.run(shots=1024)
print(res)
```
