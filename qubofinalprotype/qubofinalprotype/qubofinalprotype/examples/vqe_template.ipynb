# VQE Template
A minimal template showing how to define a parameterized circuit and evaluate an energy expectation.

```python
from qubo.circuit import QuantumCircuit
from qubo.gates import RZ, H, Measure
from qubo.simulator import StatevectorSimulator
import numpy as np
```

```python
theta = 0.3
qc = QuantumCircuit(1)
qc.add_gate('RZ', targets=[0], params=[theta])
qc.add_gate('Measure', targets=[0])
sim = StatevectorSimulator(qc)
print(sim.run(shots=1000))
```
