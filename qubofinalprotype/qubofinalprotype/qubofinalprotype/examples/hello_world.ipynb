# Qubo Quantum Simulator: Hello World
This notebook demonstrates a simple quantum circuit using the Qubo toolkit.

```python
# Import Qubo modules
from qubo.circuit import QuantumCircuit
from qubo.gates import H, X, Measure
```

```python
# Create a 1-qubit circuit and apply a Hadamard gate
qc = QuantumCircuit(1)
qc.add_gate(H(0))
qc.add_gate(Measure(0))
qc.draw()  # Visualize the circuit
```

```python
# Simulate the circuit
from qubo.simulator import StatevectorSimulator
sim = StatevectorSimulator(qc)
result = sim.run(shots=1000)
print(result)  # Show measurement statistics
```
